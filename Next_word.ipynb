```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Step 1: Create a dataset of sentences
sentences = [
"I am learning deep learning."
"Deep learning is revolutionizing AI."
"We are building a machine learning model."
"Machine learning models are powerful."
"Deep learning can solve complex problems."
"I am learning about neural networks."
"Neural networks are inspired by the brain."
"We are building a deep neural network."
"Deep neural networks have many layers."
"Neural networks can learn from data."
"I am learning about machine learning."
"Machine learning is a subset of AI."
"We are building an AI system."
"AI systems are changing the world."
"Machine learning can automate tasks."
"I am learning about data science."
"Data science involves data analysis."
"We are building a data science pipeline."
"Data science is an interdisciplinary field."
"Data science uses statistical methods."
"I am learning about data mining."
"Data mining discovers patterns in data."
"We are building a data mining algorithm."
"Data mining techniques are diverse."
"Data mining can extract valuable insights."
"I am learning about big data."
"Big data is a massive dataset."
"We are building a big data system."
"Big data requires specialized tools."
"Big data has many applications."
"I am learning about data visualization."
"Data visualization presents data graphically."
"We are building a data visualization dashboard."
"Data visualization makes data easier to understand."
"Data visualization can tell compelling stories."
"I am learning about data analysis techniques."
"Data analysis techniques are used to explore data."
"We are building a data analysis model."
"Data analysis techniques include regression and clustering."
"Data analysis can uncover hidden trends."
"I am learning about statistical inference."
"Statistical inference makes conclusions from data."
"We are building a statistical model."
"Statistical inference uses probability and statistics."
"Statistical inference can make predictions."
"I am learning about predictive modeling."
"Predictive modeling forecasts future outcomes."
"We are building a predictive model."
"Predictive modeling uses historical data."
"Predictive modeling has many real-world applications."
"I am learning about natural language processing."
"Natural language processing deals with human language."
"We are building a natural language processing system."
"Natural language processing enables machines to understand language."
"Natural language processing has many subfields."
"I am learning about computer vision."
"Computer vision enables computers to see."
"We are building a computer vision algorithm."
"Computer vision has many applications in robotics."
"Computer vision can recognize objects and faces."
"I am learning about machine learning algorithms."
"Machine learning algorithms learn from data."
"We are building a machine learning model."
"Machine learning algorithms include decision trees and support vector machines."
"Machine learning algorithms can be supervised or unsupervised."
"I am learning about deep learning models."
"Deep learning models have multiple layers."
"We are building a deep learning model."
"Deep learning models can achieve high accuracy."
"Deep learning models are used in many applications."
"I am learning about artificial intelligence."
"Artificial intelligence is the simulation of human intelligence."
"We are building an AI system."
"Artificial intelligence has the potential to transform society."
"Artificial intelligence is a rapidly growing field."
"I am learning about the history of AI."
"The history of AI is full of milestones."
"We are building a historical timeline of AI."
"The history of AI dates back to the mid-20th century."
"The history of AI is marked by periods of progress and setbacks."
"I am learning about the ethics of AI."
"The ethics of AI are important considerations."
"We are building a framework for AI ethics."
"The ethics of AI involve fairness, transparency, and accountability."
"The ethics of AI will shape the future of the field."
"I am learning about the future of AI."
"The future of AI is uncertain but exciting."
"We are building a vision for the future of AI."
"The future of AI may involve..."
]

# Step 2: Tokenize the sentences and prepare sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
total_words = len(tokenizer.word_index) + 1  # Adding 1 for padding token
# Convert sentences to sequences of integers
input_sequences = []
for sentence in sentences:
    token_list = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

# # Pad sequences so that they have the same length
max_sequence_length = max([len(x) for x in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='pre')
print(input_sequences.shape)

# # Create X (input features) and y (next word labels)
X, y = input_sequences[:,:-1], input_sequences[:,-1]

# # Step 3: Define the RNN model
model = Sequential()
model.add(Embedding(total_words, 10, input_length=max_sequence_length-1))  # Embedding layer
model.add(SimpleRNN(50, activation='relu'))  # RNN layer with 50 units
model.add(Dense(total_words, activation='softmax'))  # Output layer with total words as classes

# # Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# # Step 4: Train the model
model.fit(X, y, epochs=100, verbose=1)

# # Step 5: Test the model: Given a sentence, predict the next word
def predict_next_word(model, tokenizer, text, max_sequence_length):
#     # Tokenize the input text
    token_list = tokenizer.texts_to_sequences([text])[0]

#     # Pad the sequence to match the input shape
    token_list = pad_sequences([token_list], maxlen=max_sequence_length-1, padding='pre')

#     # Predict the next word
    predicted_probabilities = model.predict(token_list)
    predicted_word_index = np.argmax(predicted_probabilities, axis=-1)[0]

#     # Reverse the word index to get the word
    predicted_word = tokenizer.index_word[predicted_word_index]
    return predicted_word
```

```python

# Example test
input_text = "we"
predicted_word = predict_next_word(model, tokenizer, input_text, max_sequence_length)
print(f"Next word after '{input_text}': {predicted_word}")
```

```python
# Step 6: Predict the next word
def predict_next_word(seed_text, n_words, tokenizer, max_sequence_length):
    for _ in range(n_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_length-1, padding='pre')
        predicted = model.predict(token_list, verbose=0)
        predicted_word_index = np.argmax(predicted)
        for word, index in tokenizer.word_index.items():
            if index == predicted_word_index:
                seed_text += " " + word
                break
    return seed_text

# Generate a sentence
seed_text = "I am learning"
predicted_sentence = predict_next_word(seed_text, 3, tokenizer, max_sequence_length)
print("Generated Sentence:", predicted_sentence)
```
